```python
time_varying_dfs = [airquality, climate, greenery]
static_dfs = [landcover, sdi]

for df in time_varying_dfs:
    df.set_index(['COUNTYFP', 'year', 'month'], inplace=True)
    
for df in static_dfs:
        df.set_index('COUNTYFP', inplace=True)
```

```python
tvdf = pd.DataFrame()
for df in time_varying_dfs: 
    tvdf = pd.concat([tvdf, df], axis=1)
```

```python
add_cols = [col for col in tvdf.columns if '_sum' in col]
non_add_cols = [col for col in tvdf.columns if '_sum' not in col]
```

```python
tvdf.groupby(['COUNTYFP', 'year'])[add_cols].sum()
tvdf.groupby(['COUNTYFP', 'year'])[non_add_cols].mean()

yearly_tvdf = tvdf.groupby(['COUNTYFP', 'year'])[add_cols].sum().join(tvdf.groupby(['COUNTYFP', 'year'])[non_add_cols].mean())
yearly_tvdf
```
